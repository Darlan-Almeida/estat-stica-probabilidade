Importando as bibliotecas necessárias

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

17a)  grafico da Função de Densidade de Probabilidade (FDP) de uma Distribuição Normal Padrão

```python
# Parâmetros da distribuição normal (média e desvio padrão)
media = 170
desvio_padrao = 5

# Gere valores ao longo do eixo x
x = np.linspace(150 , 190 , 1000)

# Calcule a FDP para cada valor de x
pdf_values = norm.pdf(x, loc=media, scale=desvio_padrao)

# Plot da FDP
plt.plot(x, pdf_values, label='FDP da distribuição normal')
plt.title('Função de Densidade de Probabilidade (FDP) de uma Distribuição Normal Padrão')
plt.xlabel('Valores de x')
plt.ylabel('FDP(x)')
plt.legend()
plt.show()
```

Resultado da questão 17a)

```python
media = 170
desvio_padrao = 5
valor_x = 165

quantidade_alunos = 10000

probabilidade_acumulativa = 1 - norm.cdf(valor_x, loc=media, scale=desvio_padrao)


print(f"A probabilidade para x={valor_x} é: {probabilidade_acumulativa * 100:.2f}%")

alunos_acima = quantidade_alunos * probabilidade_acumulativa

print(f"A quantidade estimada de alunos acima de 165 cm é: {int(alunos_acima)}")


```

```python
media = 170
desvio_padrao = 5

# Probabilidade acumulativa alvo (75%)
probabilidade_alvo = 0.75

# Encontrar o valor x1 correspondente ao percentil 25%
x1 = norm.ppf((1 - probabilidade_alvo) / 2, loc=media, scale=desvio_padrao)

# Encontrar o valor x2 correspondente ao percentil 75%
x2 = norm.ppf(1 - (1 - probabilidade_alvo) / 2, loc=media, scale=desvio_padrao)

print(f"Intervalo simétrico em torno de 75%: ({x1:.2f}, {x2:.2f})")

# Plot da PDF
plt.plot(x, pdf_values, label='PDF da distribuição normal')

# Preencher a área correspondente ao intervalo simétrico em torno de 75%
plt.fill_between(x, pdf_values, where=[(xi >= x1) and (xi <= x2) for xi in x], color='lightblue', alpha=0.5, label='Intervalo 75%')

plt.title('Função de Densidade de Probabilidade (PDF) com Intervalo 75% Preenchido')
plt.xlabel('Valores de x')
plt.ylabel('PDF(x)')
plt.legend()
plt.show()
```
